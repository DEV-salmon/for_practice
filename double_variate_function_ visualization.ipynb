```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
from mpl_toolkits.mplot3d import Axes3D
```

```python
#사각형의 넓이(음수를 포함해야 예쁘게 나온당...)
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
x = np.arange(-5,5,0.1)
y = np.arange(-5,5,0.1)
x,y=np.meshgrid(x,y)
z=x*y
ax.plot_surface(x,y,z,cmap=cm.coolwarm)
plt.show()
```

```python
#사각형의 넓이(음수 미포함)
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
x = np.arange(0,10,0.1)
y = np.arange(0,10,0.1)
x,y=np.meshgrid(x,y)
z=x*y
ax.plot_surface(x,y,z,cmap=cm.coolwarm)
plt.show()
```

```python
#(x,y)좌표의 거리 시각화
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
x = np.arange(-5,5,0.1)
y = np.arange(-5,5,0.1)
x,y=np.meshgrid(x,y)
z=np.sqrt(x**2+y**2)
ax.plot_surface(x,y,z,cmap=cm.coolwarm)
plt.show()
```

```python
#z**2+x**2+y**2=r에서 구의 반지름이 100으로 주어졌을 때 시각화
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
x = np.arange(-5,5,0.1)
y = np.arange(-5,5,0.1)
x,y=np.meshgrid(x,y)
z=np.sqrt(100-x**2-y**2)
ax.plot_surface(x,y,z,cmap=cm.coolwarm)
plt.show()
```

```python
#z=1/tan(cos(sin(x*y))) 일때 시각화
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
x = np.arange(-5,5,0.1)
y = np.arange(-5,5,0.1)
x,y=np.meshgrid(x,y)
z=1/np.tan(np.cos(np.sin(x)*np.sin(y)))
ax.plot_surface(x,y,z,cmap=cm.coolwarm)
plt.show()
#신기해라
```
